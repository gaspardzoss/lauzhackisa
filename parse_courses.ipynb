```python
%matplotlib inline
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import scipy as sp
import urllib.parse
import re
```

```python
URL_COURSE_LIST = "https://isa.epfl.ch/imoniteur_ISAP/!GEDPUBLICREPORTS.filter"
PARAMS_COURSE_LIST = {
    'ww_b_list': 1,
    'ww_i_reportmodel': 66627699,
    'ww_i_reportModelXsl': 66627727,
}

URL_COURSE = "https://isa.epfl.ch/imoniteur_ISAP/!GEDPUBLICREPORTS.html?ww_i_reportModel=66627699&ww_i_reportModelXsl=66627723"
```

```python
def request2soup(url, params):
    r = requests.get(url, params=params)
    assert r.status_code == 200
    return BeautifulSoup(r.text, "lxml")

def create_link_from_course(url, mat):
    return url+"&"+mat

def get_mat_id_from_onclick_attr(attr):
    return attr['onclick'].split('\'')[1]
```

```python
data = request2soup(URL_COURSE_LIST, PARAMS_COURSE_LIST)
```

```python
mat_links = []
for a in data.findAll("a", { "class" : "ww_x_MAT" }):
    mat_links.append([a.text, create_link_from_course(URL_COURSE, get_mat_id_from_onclick_attr(a))])
```

```python
def get_student_list(mat_link):
    # mat_link contains the course name [0] and the link [1]
    print(mat_link[1])
    html_page = pd.read_html(mat_link[1])
    print(html_page)
```

```python
get_student_list(mat_links[400])
```
